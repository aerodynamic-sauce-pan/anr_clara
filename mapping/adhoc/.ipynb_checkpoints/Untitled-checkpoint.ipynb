```python
%matplotlib inline
imgRGB = cv.cvtColor(cv.imread('/media/mis/Disque_E/equi_script_rgb.png'), cv.COLOR_BGR2RGB)
imgDepth = cv.cvtColor(cv.imread('/media/mis/Disque_E/equi_script_depth.png'), cv.COLOR_BGR2RGB)

plt.figure(figsize=(16,16))
plt.subplot(1,2,1)
plt.imshow(imgRGB)
plt.subplot(1,2,2)
plt.imshow(imgDepth)
plt.show()

print("Min/max depth values : ", np.min(imgDepth), np.max(imgDepth))

imgRGB_clip = copy(imgRGB)
imgRGB_clip[np.where(imgDepth>200)] = 0

imgDepth_clip = copy(imgDepth)
imgDepth_clip[np.where(imgDepth>200)] = 0

plt.figure(figsize=(16,16))
plt.subplot(1,2,1)
plt.imshow(imgRGB_clip)
plt.subplot(1,2,2)
plt.imshow(imgDepth_clip)
plt.show()
```
